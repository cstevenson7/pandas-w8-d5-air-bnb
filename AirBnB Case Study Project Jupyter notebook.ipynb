```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb.groupby(['neighbourhood_group']).size()

#Answer: There are 5 neighborhood groups available and Mahattan shows up the most.
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby(['neighbourhood_group','room_type']).size()


#Answer: No the Entire home/apt is the most popular in Manhattan
```

```python
# Which hosts are the busiest and based on their reviews?

sorted_air_bnb = air_bnb.sort_values('number_of_reviews', kind = 'mergesort', ascending = False)[['host_name', 'number_of_reviews']]
sorted_air_bnb.head(5)

#Answer: Dona is the busiest based on the number of reviews
```

```python
#Which neighorhood group has the highest average price?
#print(air_bnb.describe()['price']['mean'])
air_bnb.groupby('neighbourhood_group').mean()[['price']]


#Answer = Manhattan has the highest average price
```

```python
# Which neighbor hood group has the highest total price?

air_bnb.groupby('neighbourhood_group').sum()[['price']]

#Answer = Manhattan has the highest total price
```

```python
#Which top 5 hosts have the highest total price?
air_bnb['total_price'] = air_bnb['price'] * air_bnb['minimum_nights']

sorted_air_bnb = air_bnb.sort_values('total_price', kind = 'mergesort', ascending = False)[['host_name', 'total_price']]
sorted_air_bnb.head(5)

#Answer: see below
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
zero_avail = air_bnb[air_bnb['availability_365']==0]
zero_gt = zero_avail[zero_avail['number_of_reviews'] > 100]
zero_gt
```

```python
# What host has the highest total of prices and where are they located?

#sorted_air_bnb = air_bnb.sort_values('total_price', kind = 'mergesort', ascending = False)[['host_name', 'total_price']]

sorted_air_bnb = air_bnb.sort_values('total_price', kind = 'mergesort', ascending = False)[['host_name','neighbourhood_group', 'total_price']]
sorted_air_bnb.head(5)
```

```python
# When did Danielle from Queens last receive a review?

host_Danielle = air_bnb[air_bnb['host_name'] == "Danielle"]
type(host_Danielle)


sorted_host_Danielle = host_Danielle.sort_values('last_review', kind = 'mergesort', ascending = False)[['host_name','last_review']]

sorted_host_Danielle.head(1)
```

## Further Questions
